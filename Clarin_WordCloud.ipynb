
Image-colored wordcloud
=======================


```python
from os import path
from PIL import Image
from wordcloud import WordCloud, STOPWORDS
import random
import numpy as np
import matplotlib.pyplot as plt
import os

d = path.dirname(__file__) if "__file__" in locals() else os.getcwd()

# Leer el texto

text = open(path.join(d, 'aaaa.txt')).read()

# Que lea la imágen como máscara
ln_mask = np.array(Image.open(path.join(d, "clarin.png")))

stopwords = set(STOPWORDS)

# insertamos palabras que no queremos que aparezcan
#stopwords.add("")

def color_func(word, font_size, position, orientation, random_state=None,
                   **kwargs):
    return "hsl(0, 0%%, %d%%)" % random.randint(0, 0)

wc = WordCloud(font_path='/home/thomas/Escritorio/Clarin_Wordcloud/Portada/Portada/Portada.ttf',min_font_size = 10,color_func= color_func, background_color="white", max_words=200,mask= ln_mask,
               stopwords=stopwords, contour_width=5, contour_color='red')

# Generamos el WordCloud

wc.generate(text)

# Guardado
#wc.to_file(path.join(d, "clarin_2.png"))

# Ploteado
# plt.imshow(wc, interpolation='bilinear')
plt.axis("off")
# Ajustar el tamaño acá, en el figsize.
plt.figure(figsize = (15,8))
plt.imshow(wc, cmap=plt.cm.gray, interpolation='bilinear')
plt.axis("off")
plt.show()
```
