
Image-colored wordcloud

```python
from os import path
from PIL import Image
from wordcloud import WordCloud, STOPWORDS
import random
import numpy as np
import matplotlib.pyplot as plt
import os

d = path.dirname(__file__) if "__file__" in locals() else os.getcwd()

# Leer el texto
text = open(path.join(d, 'aaaa.txt')).read()

# Que lea la imágen como máscara
lanacion_mask = np.array(Image.open(path.join(d, "/home/thomas/Escritorio/Clarin_Wordcloud/Mitre/mitre 1.jpg")))

stopwords = set(STOPWORDS)

# insertamos palabras que no queremos que aparezcan
#stopwords.add("")


def red_color_func(word, font_size, font_path, position, orientation, random_state=None):
    return "hsl(360, 100%%, %d%%)" % random.randint(30, 50)

wc = WordCloud(font_path='/home/thomas/Escritorio/Clarin_Wordcloud/Portada/Portada/Portada.ttf',min_font_size = 1,color_func= red_color_func, background_color="white", max_words=1000, mask=lanacion_mask,
               stopwords=stopwords, contour_width=3, contour_color='white', width=lanacion_mask.shape[1],
               height=lanacion_mask.shape[0])

# Generamos el WordCloud
wc.generate(text)

# Guardado
#wc.to_file(path.join(d, "clarin.png"))

# Ploteado
#plt.imshow(wc, interpolation='bilinear')
plt.axis("off")
# Ajustar el tamaño acá, en el figsize.
plt.figure(figsize = (25,15))
plt.imshow(wc, cmap=plt.cm.gray, interpolation='none')
plt.axis("off")
plt.show()
```
